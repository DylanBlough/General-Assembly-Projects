## Cleaning and Preprocessing

```python
# Remove outliers

indexNames = train[train['Gr Liv Area'] > 4000].index
train.drop(indexNames, inplace = True)
```

```python
# reformatting the columns to all lowercase no space

def column_rename(dataframe):
    dataframe.columns = [x.lower().replace(' ','_') for x in dataframe.columns]
```

```python
column_rename(train)
```

```python
column_rename(test)
```

```python
# checking the columns for null data values

train.isnull().sum()
```

```python
# looks like theres a lot of columns, written as function instead

def check_cols(df):
    check_cols = df.select_dtypes([np.object]).columns # telling it to only look at obj columns
    for col in check_cols:
        n_nan = df[col].isnull().sum()
        if n_nan > 0:
            print(f"{col} has {n_nan} NaNs")
    return

# I originally had an else statement to print results for 0 NaN columns, but it was difficult to read and
# we're only concerned with NaN columns
```

```python
check_cols(train)
```

```python
# let's start filling values

def replace_null(df, col, fillval='None'):
    
    df.loc[df[col].isnull(), col] = fillval
    
    return df
```

```python
def replace_na(df, col, fillval = 0):
    df.loc[df[col].isna(), col] = fillval
    return df
```

```python
# features that we're amending or dropping

none_col = ['garage_type','garage_finish','garage_qual','garage_cond','fence',
                  'fireplace_qu','bsmtfin_type_2','bsmtfin_type_1','bsmt_exposure',
                  'bsmt_cond','bsmt_qual','mas_vnr_type','misc_feature']
zero_col = ['alley','lot_frontage','mas_vnr_area', 'garage_cars', 'garage_area']
drop_rows = ['electrical']
drop_col = ['pool_qc']
```

```python
# run the functions

for col in none_col:
       train = replace_null(train, col, fillval='None')
for col in zero_col:
        train = replace_null(train, col, fillval=0.)
        train = replace_na(train, col, fillval=0.)
for col in drop_rows:
     train = train[~train[col].isnull()]
for col in drop_col:
       train.drop(col, axis=1, inplace=True)
```

```python
# check the results

train.isnull().sum()
```

## Dummying Categorical Variables

```python
# since the geographic features seem to have a strong correlation, let's dummy out the neighborhood column

hood_dummy = pd.get_dummies(train['neighborhood'])
```

```python
hood_dummy.head(5)
```

```python
# merge the dummies back in

train = train.merge(hood_dummy, left_on = 'id', right_on = 'Blmngtn', how = 'left')

# apply it to our test set too

hood_dummy_test = pd.get_dummies(test['neighborhood'])
test = test.merge(hood_dummy_test, left_index = True, right_index = True)
```

```python
# cleaning the dummy columns 

zero_col = ['Blmngtn', 'Blueste', 'BrDale',
            'BrkSide', 'ClearCr', 'CollgCr', 'Crawfor', 'Edwards', 'Gilbert', 'IDOTRR', 'MeadowV', 'Mitchel',
            'NoRidge', 'NPkVill', 'NridgHt', 'NWAmes', 'OldTown', 'SWISU', 'Sawyer', 'SawyerW', 
            'Somerst', 'StoneBr', 'Timber', 'Veenker']

for col in zero_col:
        train = replace_null(train, col, fillval=0.)
        train = replace_na(train, col, fillval=0.)
```
